```python
#Importing dependencies 
import pandas as pd
```

```python
# reading in csv's and validating all data was captured in the converstion from xml to csv 
# comment one of these out at a time to see both views of the data
df_2017 = pd.read_csv('1.1.2017.csv')
df_2017[['Unique\nEmployee ID #']] = df_2017[['Unique\nEmployee ID #']].fillna(value=0)
df_2017['Unique\nEmployee ID #'] = df_2017['Unique\nEmployee ID #'].astype('int64') 
df_2017['Unique\nEmployee ID #'] = df_2017['Unique\nEmployee ID #'].replace(' ', '')
df_2017.head()
```

```python
# reading in csv's and validating all data was captured in the converstion from xml to csv 
# filling na values with 0 
# casting the employee id to int64 because it was being cast as sicietific notation 
df_2018 = pd.read_csv('1.1.2018.csv')
df_2018[['Unique\nEmployee ID #']] = df_2018[['Unique\nEmployee ID #']].fillna(value=0)
df_2018['Unique\nEmployee ID #'] = df_2018['Unique\nEmployee ID #'].astype('int64') 
df_2018['Unique\nEmployee ID #'] = df_2018['Unique\nEmployee ID #'].replace(' ', '')
df_2018['Date of Termination'] = df_2018['Date of Termination'].replace('#VALUE!', 'N/A')
df_2018['Retirement after Hire'] = df_2018['Retirement after Hire'].replace('#VALUE!', 'N/A')
df_2018['Termination after Hire'] = df_2018['Termination after Hire'].replace('#VALUE!', 'N/A')
df_2018 = df_2018.drop(45)
df_2018.head()
```

```python
# merging data sets 
common = df_2018.merge(df_2017,on=['Unique\nEmployee ID #','Unique\nEmployee ID #'])
common.head()
```

```python
# use isin to see if employee ID is in only the 2018 dataset. These emplyees should be the new hires since their is no record of them in the provious data set. 
new_hires_df = df_2018[(~df_2018['Unique\nEmployee ID #'].isin(common['Unique\nEmployee ID #']))&(~df_2018['Unique\nEmployee ID #'].isin(common['Unique\nEmployee ID #']))]
new_hires_df
```

```python
# emplyees that have been terminated in 2017 
# these emplyees shouldnt be in the 2018 data set 
terminated = df_2017.loc[:,['Name','Date of Termination']]

# terminated 
cleaned_terminated = terminated[terminated['Date of Termination'].notna()]
cleaned_terminated
```

```python
# checking to see if the emplyees terminated in 2017 are shown as terminated in the 2018 data set as well 

df_2018.loc[df_2018['Name'].isin(['SCOTT W.','JEFFREY W.', 'KEITH W.', 'KIMBERLY W.','JANNA W.', 'JESSE W.', 'JEFFREY W.', 'CHAD W.', 'RYAN Y.'])]
```

```python
# emplyees that have been terminated in 2018 
# these emplyees should have record of this termination in 2017
terminated_2018 = df_2018.loc[:,['Name','Date of Termination']]

# terminated 
cleaned_terminated_2018 = terminated_2018[terminated_2018['Date of Termination'].notna()]
cleaned_terminated_2018
```

```python
# checking to see if the emplyees that show a record of being terminated in 2018 are shown as terminated in the 2017 data set as well 
terminated_emplyees = df_2017.loc[df_2017['Name'].isin(['BRIAN T.','SCOTT W.','JEFFREY W.', 'KEITH W.', 'KIMBERLY W.','JANNA W.','JEFFREY W.', 'JESSE W.', 'CHAD W.', 'RYAN Y.'])]
terminated_emplyees
```

```python
# this emplyee is shown as Active in the 2017 data set regardless of the fact the 2018 data set shows him as terminated on 7/1/15
terminated_emplyees[terminated_emplyees['Status'].str.contains('Active')]
```

```python
# checking retirement date in 2018 data set 
retired = df_2018.loc[:,['Name','Date of Retirement']]
retired_2018 = retired[retired['Date of Retirement'].notna()]
retired_2018
```

```python
# checking retirement date in 2017 data set 
retired = df_2017.loc[:,['Name','Date of Retirement']]
retired_2017 = retired[retired['Date of Retirement'].notna()]
retired_2017
```
